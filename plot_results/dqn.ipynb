```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
dqn_win_rate = np.array([0.2115,0.1043,0.07356,0.02825,0.01544,6.8549e-3,0.00624,0.0034,0.00213,0.001,0.0076,0.00109,0,0,0,0,0])
dqn_lengths = np.array([i for i in range(4, 21)])
dqn_g_win_rate = np.array([0.9486,0.9545,0.9638,0.9689,0,0,0,0,0])
dqn_g_lengths = np.array([i for i in range(4, 22, 2)])
dqn_her_win_rate = np.array([0.9499,0.9542,0.9597,0.9717,0.981,0.9826,0.9872,0.9865,0.994,0,0,0,0,0])
dqn_her_lengths = np.array([i for i in range(4, 32, 2)])
figure = plt.figure()
plt.plot(dqn_lengths, dqn_win_rate, label="dqn")
plt.plot(dqn_g_lengths, dqn_g_win_rate, label="dqn_with_goal")
plt.plot(dqn_her_lengths, dqn_her_win_rate, label="dqn_with_her")
plt.grid()
# plt.ylim([0,1.1])
# plt.title("DQN")
plt.ylabel("win rate")
plt.xlabel("nbits")
plt.legend()
# plt.show()
plt.savefig("length.pdf")
```

```python
win_rate = np.array([0.2115,0.1043,0.07356,0.02825,0.01544,6.8549e-3,0.00624,0.0034,0.00213,0.001,0.0076,0.00109,0,0,0,0,0])
lengths = np.array([i for i in range(4, 21)])
figure = plt.figure()
plt.plot(lengths, win_rate)
plt.grid()
# plt.ylim([0,1.1])
plt.title("DQN")
plt.ylabel("win rate")
plt.xlabel("nbits")
# plt.show()
plt.savefig("dqn_length.pdf")
```
